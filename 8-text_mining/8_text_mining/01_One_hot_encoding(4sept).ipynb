```python
import pandas as pd
```

```python
df=pd.read_csv("homeprices.csv")
df#read the file
```

# Using Pandas to Create dummy variables

```python
dummies=pd.get_dummies(df.town)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
final=merged.drop(['town'],axis='columns')
final
```

# Dummy Vairable Trap

When you can derive 1 variable from other variables,they are known to be multi-collinear.so to avoid this we will drop one column 


When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.
NOTE: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ou are using is not handling this for you

```python
final=final.drop(['west windsor'],axis='columns')
final
```

```python
X=final.drop('price',axis='columns')
X#the input columns is "X" always
```

```python
y=final.price#the target column is "y" always
```

```python
y#so our target column is price
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
#here it is going to define the model
```

```python
#here it is going to train the model
model.fit(X,y)
```

```python
model.predict(X)#2600 sqr ft home in new jersey
#this is going to predict the results
```

```python
model.score(X,y)#this is going to give us the accuracy of the results
```

```python
model.predict([[3400,0,0]])#3400 sqr ft home in west windsor
#here 0,0 stands for west windsor
```

```python
model.predict([[2800,0,1]])
#2800 sqr home in robbinville
#0,1 stands for robbinville
```

# Using sklearn OneHotEncoder(5sept)

#first step is to use label encoder to convert town names into numbers

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town','area']].values
X
```

```python
y = dfle.price.values
y
```

```python
#now use one hot encoder to create dummy variables for each of the town
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])],remainder = 'passthrough')
#The code creates a ColumnTransformer that one-hot encodes
#the first column #(index 0) of the dataset and leaves
#the remaining columns unchanged (passthrough)
#this is useful when you have a mix of categorical and numerical features 
#in your dataset and want to apply specific transformations
#only to certain columns while keeping others as they are.
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]#droping first column using slicing
```

```python
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,3400]])
#3400sqr ft home in west windsor
```

```python
model.predict([[1,0,2800]])#2800 sqr ft home in robbinsville
```

```python
model.predict([[3400,0,1]])
```
