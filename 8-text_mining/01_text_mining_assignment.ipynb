```python
#lets extract the review-title now just with ratings now
text1=[]
for i in range(0,len(title3)):
    text1.append(title3[i].get_text())
print(len(text1))
text1
```

```python
rate1=[]
for i in range(len(text1)):
    rating1=text1[i].split(' ')
    rate1.append(rating1[0])
print(rate1)
print(len(rate1))
```

```python
reviews=[]
for i in range(0,len(text1)):
    reviews.append(text1[i].get_text())
```

```python
text11=[]
text12=[]
for i in range(0,len(text1)):
    text11 = [t.split("\n") for t in text1]
    text12.append(text11.get_text())
text11
```

```python
text12 = [] 
for t in text1:
    text11 = t.split("\n")
    text12.extend(text11) 
print(text12)
```

```python
reviews = [text12[i] for i in range(1,len(text12),3)] 
print(reviews)
```

```python
#lets create the file in .csv file name
import pandas as pd
df=pd.DataFrame()
df['index']=range(1,9)
df['Review']=reviews
df['rate']=rate1

df
```

```python
from textblob import TextBlob
df['polarity']=df['Review'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
df['polarity']
```
