```python
#from this we get to know that outliers are present
#for removing outliers we will use IQR method
```

```python
def detect_outliers_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    return (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))
```

```python
# Detecting outliers in multiple columns
outliers = detect_outliers_iqr(df[['cet_cuttoff','Fees']])
print(outliers)
# Remove rows where any column has an outlier
df_cleaned = df[~outliers.any(axis=1)]
print(df_cleaned)
```

```python
#Normalize the data
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return x

df_norm = norm_func(df_cleaned.iloc[:, [2, 5]])

df_norm
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
#identifying the ideal number of clusters using TWSS and elbow curve
TWSS=[]
k=list(range(2,8))
for i in k:
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)

    TWSS.append(kmeans.inertia_)#total sum of square

TWSS
```

```python
plt.plot(k,TWSS,'ro-');
plt.xlabel("No_of_clusters");
plt.xlabel("Total_within_SS")
```

```python
#Preprocessing using Min Max Scaler
scaler=MinMaxScaler()
scaler.fit(df[['cet_cuttoff']])
df['cet_cuttoff']=scaler.transform(df[['cet_cuttoff']])
```

```python
scaler.fit(df[['Fees']])
df['Fees']=scaler.transform(df[['Fees']])

plt.scatter(df.Fees,df['cet_cuttoff'])
df.head()

plt.scatter(df.Fees,df['cet_cuttoff'])
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['Fees','cet_cuttoff']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Fees,df1['cet_cuttoff'],color='green')
plt.scatter(df2.Fees,df2['cet_cuttoff'],color='red')
plt.scatter(df3.Fees,df3['cet_cuttoff'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',s=200,label='centroid')
plt.xlabel('Fees')
plt.ylabel('cet_cuttoff')
plt.legend()
```
